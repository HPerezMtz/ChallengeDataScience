### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



## ***En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.***

```python
ingreso_total_tienda1 = tienda['Precio'].sum()
print(f"Ingreso total de Tienda 1: {ingreso_total_tienda1}")
```

```python
ingreso_total_tienda2 = tienda2['Precio'].sum()
print(f"Ingreso total de Tienda 2: {ingreso_total_tienda2}")
```

```python
ingreso_total_tienda3 = tienda3['Precio'].sum()
print(f"Ingreso total de Tienda 3: {ingreso_total_tienda3}")
```

```python
ingreso_total_tienda4 = tienda4['Precio'].sum()
print(f"Ingreso total de Tienda 4: {ingreso_total_tienda4}")
```

```python
import matplotlib.pyplot as plt

ingresos_totales = {
    'Tienda 1': ingreso_total_tienda1,
    'Tienda 2': ingreso_total_tienda2,
    'Tienda 3': ingreso_total_tienda3,
    'Tienda 4': ingreso_total_tienda4
}

tiendas = list(ingresos_totales.keys())
ingresos = list(ingresos_totales.values())

plt.figure(figsize=(8, 5))
plt.plot(tiendas, ingresos, marker='o', linestyle='-', color='blue')
plt.xlabel("Tienda")
plt.ylabel("Ingreso Total")
plt.title("Ingreso Total por Tienda")
plt.grid(True)
plt.show()
```

### Análisis de la Tienda 4 (menor rendimiento)

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
promedios_calificacion = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

# Normalize the data for comparison on the same scale
# Normalizing revenues
max_ingreso = max(ingresos)
min_ingreso = min(ingresos)
normalized_ingresos = [(i - min_ingreso) / (max_ingreso - min_ingreso) for i in ingresos]

# Normalizing average ratings
max_calificacion = max(promedios_calificacion)
min_calificacion = min(promedios_calificacion)
normalized_calificaciones = [(c - min_calificacion) / (max_calificacion - min_calificacion) for c in promedios_calificacion]


x = np.arange(len(tiendas))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, normalized_ingresos, width, label='Ingreso Total (Normalizado)')
rects2 = ax.bar(x + width/2, normalized_calificaciones, width, label='Calificación Promedio (Normalizado)')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Valor Normalizado')
ax.set_title('Comparación de Ingreso Total y Calificación Promedio por Tienda')
ax.set_xticks(x)
ax.set_xticklabels(tiendas)
ax.legend()

# Add labels on top of bars
def autolabel(rects, data):
    for i, rect in enumerate(rects):
        height = rect.get_height()
        ax.annotate(f'{data[i]:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1, ingresos)
autolabel(rects2, promedios_calificacion)


plt.show()
```
